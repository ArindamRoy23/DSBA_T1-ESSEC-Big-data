```python
import numpy as np
import pandas as pd

a = "Hello my name is something "


b = "Hello, my name is actually something"

def shingle_it(a):
    shingles = []
    for i in range(len(a)):
        if i <len(a) - 3:
            shingles.append(a[i:i+3])
    return (shingles)



shi_a = shingle_it(a)
shi_b = shingle_it(b)


shi_a_df = pd.DataFrame()
shi_b_df = pd.DataFrame()
```

```python
shi_a_df['A'] = np.ones((len(shi_a)))
shi_a_df['I'] = shi_a
shi_b_df['B'] = np.ones((len(shi_b)))
shi_b_df['I'] = shi_b
```
